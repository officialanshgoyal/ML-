### Categories
1. Categorical Plots (hue can be used | comparison with categorical )
   a. countplot  (single para )
   b.barplot  ( one parameter categorical and one par numercial)
   c.boxplot
   d.violinplot
   e.swarmplot

2. Distribution Plots 
   a.distplot
   b.displot
   c.jointplot
   d.pairplot
   e.kdeplot
   f.rugplot
   g.scatterplot
   
3. Matrix Plot
   a. heatmap
   b. clustermap
   c. pairplot( same as above one)

4. Regression plot
   a. lmplot
   b. regplot

   a.PairGrid

```python
sns.countplot(x='day',data=tips)
```

```python

sns.countplot(x='day',data=tips, hue= 'sex')
```

```python
sns.barplot(x='day',y='total_bill',data= tips)
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day', y='total_bill',data= tips,estimator=sum)

```

```python
sns.barplot(x='day', y='total_bill',data= tips, hue='time' ,estimator=np.mean)
```

```python
sns.boxplot(x='day', y='total_bill',data= tips, hue='smoker' )
```

```python
sns.violinplot(x='day', y='total_bill',data= tips, hue='smoker' )
```

```python
sns.swarmplot(x='day', y='total_bill',data= tips, hue='smoker' )
```
