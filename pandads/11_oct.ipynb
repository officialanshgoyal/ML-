```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Types of Graphs
# 1. plot -          line graph , between two parameterw
# 2. bar  -          bar graph of single parameter
# 3. hist -          histogram of a branch of data i.e frequency of single param
# 4. scatter -       dot graph between two params 
# 5. boxplot / box - statstical overview in form of box for a single param mean, outliers,quantiles(25% , 75% ), IQR
#                     ( Optional - catagorical data )
# 6.stem -           more like bar graph , Lines instead of bars 
# 7.Violimplot -     Violin shaped statistical overview graph
#                    spread of data (more liike hist)
#                    single parameter (optional cat - data)
```

```python
x = np.linspace(0,5,11)
x1 =np.arange(1,11,3)
```

```python
x
```

```python
x1
```

```python
y= x**3
```

```python
plt.plot(x,y)
plt.title('plot graph')


plt.show()
```

```python
plt.bar(x,y)
plt.title('bar graph')
plt.show()
```

```python
plt.hist(y)
plt.title('hist graph')
```

```python
plt.scatter(x,y)
plt.title('scatter graph')
plt.grid()
plt.xlabel('x -- > ')
plt.ylabel(' y -- >  ')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()

```

```python
plt.boxplot(x)
```

```python
plt.stem(x,y)
```

```python
plt.violinplot(x)
```

```python
y2=x**4
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
fig =plt.figure(figsize=(10,5))
plt.plot(x,y, lw=3, marker='o' , markersize=15,markerfacecolor = 'yellow')
plt.plot(x,y2, marker = 'v' ,markersize= 10)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```
