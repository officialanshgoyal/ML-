```python
import numpy as np
import pandas as pd 
```

```python
data = {'country':['India','USA','Japan','China','India','USA'],
        'Year':['2010','2010','2010','2015','2015','2015'],
        'Continent':['Asia','NA','Asia','Asia','Asia','NA'],
        'Value':[20,10,40,50,60,55]
              
       } 
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
pd.pivot_table(df,index=['Year'],aggfunc='sum')
```

```python
pd.pivot_table(df,index=['country'],aggfunc='sum')
```

```python
pd.pivot_table(df,index=['Continent'],aggfunc='sum')
```

## Concat

```python
people1={'id':[1,2,3,4],
         'name':['ab','rf','rt','ed'],
         'age':[45,25,41,30]
        }        
                 
                 
people2={'id':[11,22,33,44],
         'name':['awb','rgf','kak','ejd'],
         'age':[35,15,40,35]
        }        
                                
                 
                 
people3={'id':[111,222,333,444],
         'name':['oab','rnf','knk','ecd'],
         'age':[55,75,81,50]
        }        
                                
                 
                 
                 
                 
                 
                 
                 
                 
                 
                 
                 
                 
                 
                 
                 
```

```python
df1=pd.DataFrame(people1)
df2=pd.DataFrame(people2)
df3=pd.DataFrame(people3)




```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2,df3],ignore_index = True)
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
people4={'id':[1,2,3,4],
         'name':['ab','rf','rt','ed'],
         'age':[45,25,41,30]
        }        
                 
                 
people5={'id':[1,2,3,4 ],
         'name':['ab','rf','rt','ed'],
         'sales':[535,415,540,235]
        }        
                          
```

```python
df4=pd.DataFrame(people4)
df5=pd.DataFrame(people5)
```

```python
df4
```

```python
df5
```

```python
pd.merge(df4,df5,on=['id','name'])
```

```python
pd.merge(df4,df5,on=['id',])
```

```python
pd.merge(df4,df5,)
```

```python
people6={'id':[1,2,3,4],
         'name':['ab','rf','rt','ed'],
         'age':[45,25,41,30]
        }        
                 
                 
people7={
         's_name':['ab','rf','rt','ed'],
         'ph':[34335,466515,576540,3286235]
        }       
```

```python
df6=pd.DataFrame(people6)
df7=pd.DataFrame(people7)
```

```python
df6
```

```python
df7
```

```python
df6.join(df7)
```
